```python

# Entferne Zeilen, bei denen das Land fehlt (NaN)
df_cleaned = df.dropna(subset=['country'])

# Teile die Länder auf, falls mehrere Länder in einem Eintrag sind
all_countries = df_cleaned['country'].str.split(', ', expand=True).stack()

# Zähle die Häufigkeit jedes Landes
country_counts = all_countries.value_counts()

# Finde das Land mit der meisten Häufigkeit
most_country_producedmovie = country_counts.idxmax()
most_country_producedmovie_count = country_counts.max()

# Ausgabe des Ergebnisses
print(f"Das Land, welches am meisten Filme & Serien produziert hat, heißt: {most_country_producedmovie}")
print(f"Es hat {most_country_producedmovie_count} Filme & Serien produziert.")
```

```python

#Top 10 Länder nach Produktion
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Sortiere die Länder nach den Produktionen (absteigend) und wähle die Top 10 aus
top_countries = country_counts.sort_values(ascending=False).head(10).reset_index()
# Ändern der Spaltennamen
top_countries.columns = ['Land', 'Anzahl Produktionen']

# Seaborn-Balkendiagramm erstellen
plt.figure(figsize=(10, 6))  # Größe des Diagramms
sns.barplot(x='Land', y='Anzahl Produktionen', data=top_countries, palette='viridis')  # Balkendiagramm
plt.title('Top 10 Länder nach Produktion', fontsize=16)  # Titel
plt.xlabel('Land', fontsize=14)  # X-Achsenbeschriftung
plt.ylabel('Anzahl Produktionen', fontsize=14)  # Y-Achsenbeschriftung
plt.xticks(rotation=45)  # Drehe die Länderbezeichnungen für bessere Lesbarkeit
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Gitterlinien für die Y-Achse
plt.tight_layout()  # Bessere Anpassung des Layouts
plt.show()  # Diagramm anzeigen
```
